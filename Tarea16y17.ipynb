```julia
function espaciofasex1(a,b,alpha)
    h=0.0001
    X=[]
    x=a
    V=[]
    T=[]
    t=0
    v=b
    while t<=alpha
        push!(X,x)
        push!(V,v)
        push!(T,t)
        x=(a*cos(t)*(e^t))+(b*(e^t)*sin(t))
        v=(-a*(e^t)*sin(t))+(b*(e^t)*cos(t))
        t=t+h
    end
    plot(X,V,color="blue")
end
    
    
```

```julia
function espaciofasex2(a,b,alpha)
    h=0.0001
    X=[]
    x=b
    V=[]
    T=[]
    t=0
    v=-a
    while t<=alpha
        push!(X,x)
        push!(V,v)
        push!(T,t)
        x=(b*cos(t)*(e^t))+(-a*(e^t)*sin(t))
        v=(-b*(e^t)*sin(t))+(-a*(e^t)*cos(t))
        t=t+h
    end
    plot(X,V,color="red")
end
    
```

```julia
function EFase(alpha)
    a=rand(1:100)
    b=rand(1:100)
    espaciofasex2(a,b,alpha)
    espaciofasex1(a,b,alpha)
end
```
