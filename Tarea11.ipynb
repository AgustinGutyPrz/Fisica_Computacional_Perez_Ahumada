__Ejercicio 1:__. Implementar método implicito de Euler como quieran.

```julia
function eulerMethod(f,alfa,h,k)
    T = []
    W = []
    t=0
    w = alfa
    j=(8/h)+1
    while t<=8
        push!(T,t)
        push!(W,w)
        s=f(k,w)
        w = w + h*s
        t=t+h
    end  
    xs = linspace(0.0,8.0,j)
ys = map(x -> alfa*exp(-2.5*x), xs)
    plot(xs,ys,color="red")
    scatter(T,W,color="green")
end
```

```julia
function h(k,w)
    return -k*w
end
```


__Ejercicio 3:__.Implementar método de Runge-Kutta de orden 4.

```julia
function MetodoRunge(f,alfa,h) #función para la aproximación mediante el método runge de cuarto grado
    U = []
    V = []
    u=0
    v = alfa
    while u<=8
        push!(U,u)
        push!(V,v)
        k1=(f(u,v)) #K1,k2,k3,k4 son las fórmulas para calcular las constantes para el método
        k2=(f(u+(h/2),v+(h*k1/2)))#runge de 4 grado.
        k3=f(u+(h/2),v+(h*k2/2))
        k4=f(u+h,v+(h*k3))
        u = u + h #son los pasos de tiempo y es válido ya que son constantes.
        v = v + (h/6)*(k1+(2(k2+k3))+k4) #Es según la fórmula, una aproximación del método de runge
    end
    scatter(U,V,color="gold")
end
```

```julia
function comparacionNewtRung(f,alfa,h,k)
    MetodoRunge(f,alfa,h)
    eulerMethod(f,alfa,h,k)
end
```
