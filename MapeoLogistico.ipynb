```julia
function lyapunov(x,u,N)                 
    ly=0    
for i in 1:N                                 
    x = u*x*(1-x)                       
    ly = ly+log(abs(u*(1-2*x)))                 
    end
    return ly/N                              
    end
```

```julia
using PyPlot
```

```julia
conj=linspace(0,1,1000)
cnjt=linspace(1,5,1000)
X =[]
Y =[]
for i in 1:length(cnjt)
    push!(X,bifurcacion.(conj,cnjt[i],1000))
    push!(Y,ones(length(conj))*cnjt[i])
    end
```

```julia
plot(cnjt,lyapunov.(0.5,cnjt,1000),"o",markersize=0.5)
title("Lyapunov")
grid()
```

```julia
plot(Y,X,"o",markersize=0.05)
title("Bifurcaci√≥n")
grid()
```

```julia
plot(Y,X,"o",markersize=0.05)
plot(cnjt,lyapunov.(0.5,cnjt,1000),"o",markersize=0.5)
title("Bifurcaci√≥n y Lyapunov")
grid()
```
